```python
#import dependencies
import pandas as pd
import matplotlib as plt
```

```python
# read csv file downloaded from open data sfgov site
csv_path = "Crimes_-_2001_to_present.csv"
data = pd.read_csv(csv_path)

#explore data 
data.head(100)
```

```python
#display column headers
data.columns
```

```python
# create a reduced dataframe with only the columns relevant
df = data[['Primary Type', 'Description',  'Date', 'Year', 'ID', 'Community Area']]
df.head()
```

```python
df['Primary Type'].value_counts().head()
```

```python
#Filter the dataframe for Larceny/Theft incidents only
df_ld = df.loc[df["Primary Type"] == "THEFT"]
df_ld

#Save Larceny/Theft data to a new csv
df_ld.to_csv('data larceny and theft with year label.csv')
```

```python
df_ld
```

```python
# Visualization # 1. Line chart to show trend of larceny/theft over time in the entire dataset.
import matplotlib.pyplot as plt
csv_path = "df_ld.csv"
data_sf_ld = pd.read_csv(csv_path)
df_ld.Year.value_counts().sort_index().plot(figsize =(15,15))
data_sf_ld.Year.value_counts().sort_index().plot(figsize =(15,15))
plt.show()
```

```python
df_ld_clean = df_ld.loc[(df_ld['Year'] <= 2017) & (df_ld['Year'] >= 2003)]
df_ld_clean.Year.value_counts().sort_index()
```

```python
# Create a new dataframe df2 which has counts by district and Year
df_pd = df_ld.groupby(['Year', "Community Area"]).count()
df2 = pd.DataFrame(df_pd['Primary Type'])
df2.head()
```

```python
# Reshape the dataframe to a pivot table for easy visualization
table = pd.pivot_table(df2, index='Year',columns='Community Area')
table.head()
```

```python
#Visualization # 2 Display data in table as a multiple line chart with a line for each neighborhood to plot incodents over time
table.plot(figsize = (15,15))
```

```python
#Visualization # 3 Bar Chart by neighborhood 
table.plot(kind="bar", stacked = True, figsize = (15,15), rot =0)

plt.show()
```

```python
# Parse 'Descript' field to tag each incident in larcent/theft dataframe as either a'petty' or 'grand' theft based on a string wildcard 
df_ld.Description.value_counts()
```

```python
df_test300 = df_ld
df_test300.Description.value_counts()
```

```python
def change(x):
    if 'UNDER'in x:
        return 'Petty Theft'
    elif 'OVER' in x:
        return 'Grand Theft'
    else:
        return 'Other'
    
    
df_ld['Description'] = df_ld['Description'].apply(change)
```

```python
#Updated dataframe with new values for 'Descript' columns
df_ld.Description.value_counts()
```

```python
#Visualization # 5 Line chart over time for petty theft vs grand theft incidents
table2 = pd.pivot_table(df_ld, index='Year',columns='Description')
table2.plot()
```
