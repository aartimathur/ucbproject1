```python
import pandas as pd
```

```python
csv_path = "Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv"
data = pd.read_csv(csv_path)
data.head(100)
```

```python
data.columns
```

```python
df = data[['Category', 'Descript', 'DayOfWeek', 'Date', 'Time', 'PdDistrict', 'X', 'Y', 'PdId']]
df.head()
```

```python
df['Date'] = pd.to_datetime(df.Date)
df.head()
```

```python
df['Year'] = df.Date.dt.year
```

```python
df
```

```python
df_ld = df.loc[df["Category"] == "LARCENY/THEFT"]
df_ld
df_ld.to_csv('data larceny and theft with year label.csv')
```

```python
import matplotlib.pyplot as plt
df_ld.Year.value_counts().sort_index().plot()
plt.show()
```

```python
ts_07 = pd.to_datetime("1/1/2007")
ts_09 = pd.to_datetime("1/1/2010")
df_0709 = df_ld.loc[df_ld.Date >= ts_07, :]
df_0709 = df_0709.loc[df_ld.Date < ts_09, :]

df_0709.to_csv('data from 07 to 09.csv')
df_0709.head()
```
