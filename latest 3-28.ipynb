```python
#import dependencies
import pandas as pd
import matplotlib as plt
```

```python
# read csv file downloaded from open data sfgov site
csv_path = "Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv"
data = pd.read_csv(csv_path)

#explore data 
data.head(100)
```

```python
#display column headers
data.columns
```

```python
# create a reduced dataframe with only the columns relevant
df = data[['Category', 'Descript', 'DayOfWeek', 'Date', 'Time', 'PdDistrict', 'X', 'Y', 'PdId']]
df.head()
```

```python
# Parse the 'Date' column for year and create a new column for Year
df['Date'] = pd.to_datetime(df.Date)
df['Year'] = df.Date.dt.year

df.head()
```

```python
#drop 2018 records
df = df[df.Year!=2018]
```

```python
#Filter the dataframe for Larceny/Theft incidents only
df_ld = df.loc[df["Category"] == "LARCENY/THEFT"]
df_ld

#Save Larceny/Theft data to a new csv
df_ld.to_csv('data larceny and theft with year label.csv')
```

```python
# Visualization # 1. Line chart to show trend of larceny/theft over time in the entire dataset.
import matplotlib.pyplot as plt
df_ld.Year.value_counts().sort_index().plot(figsize =(15,5))
plt.xlabel("Year of Incident")
plt.ylabel("Number of Incidents")
plt.title("Fig 1: Larceny/Theft incidents in SF, trend Over Time")
plt.vlines(2010,0,50000, linestyles='dashed')
plt.show()
plt.savefig("Fig1_TheftbyYear.png")
```

```python
#Split the dataset three years before and after 2010 and get aggregate values.
ts_07 = pd.to_datetime("1/1/2007")
ts_09 = pd.to_datetime("1/1/2010")
df_0709 = df_ld.loc[df_ld.Date >= ts_07, :]
df_0709 = df_0709.loc[df_ld.Date < ts_09, :]

df_0709.to_csv('data from 07 to 09.csv')
df_0709.head()
```

```python
# Create a new dataframe df2 which has counts by district and Year
df_pd = df_ld.groupby(['Year', "PdDistrict"]).count()
df2 = pd.DataFrame(df_pd['Category'])
df2.head()
```

```python
# Reshape the dataframe to a pivot table for easy visualization
table = pd.pivot_table(df2, index='Year',columns='PdDistrict')
table.head()
```

```python
#Visualization # 2 Display data in table as a multiple line chart with a line for each neighborhood to plot incodents over time
table.plot(figsize = (15,8))
plt.xlabel("Year of Incident")
plt.ylabel("Number of Incidents")
plt.vlines(2010,0,15000, linestyles='dashed')
plt.title("Fig 4: Larceny/Theft incidents by SF Neighborhood, Trend Over Time")
plt.savefig("Fig4_TheftbyDistrict.png")
```

```python
#Visualization # 3 Bar Chart by neighborhood 
table.plot(kind="bar", stacked = True, figsize = (15,8), rot =0)
plt.xlabel("Year of Incident")
plt.ylabel("Number of Incidents")
plt.title("Fig 3: Larceny/Theft incidents by SF Neighborhood")
plt.savefig("Fig3_TheftbyDistrict.png")
```

```python
# Parse 'Descript' field to tag each incident in larcent/theft dataframe as either a'petty' or 'grand' theft based on a string wildcard 
df_ld['Descript'].head()
```

```python
def change(x):
    if 'PETTY THEFT'in x:
        return 'Petty Theft'
    elif 'GRAND THEFT' in x:
        return 'Grand Theft'
    else:
        return 'Other'
    
    
df_ld['Descript'] = df_ld['Descript'].apply(change)
```

```python
#Updated dataframe with new values for 'Descript' columns
df_ld.head()
```

```python
#Visualization # 5 Line chart over time for petty theft vs grand theft incidents
reduced_df = df_ld.groupby(['Year', 'Descript']).count()['Date']

df4 = reduced_df.reset_index()

df_petty = df4.loc[df4['Descript'] == 'Petty Theft']
df_grand = df4.loc[df4['Descript'] == 'Grand Theft']
df_other = df4.loc[df4['Descript'] == 'Other']

plt.figure(figsize = (15,8))
plt.xlabel("Year of Incident")
plt.ylabel("Number of Incidents")
plt.title("Fig 2: Grand Theft vs Petty Theft incidents, Trend Over Time")
plt.vlines(2010,0,35000, linestyles='dashed')

petty, = plt.plot(df_petty.Year, df_petty.Date, c='b', label = 'Petty')
grand, = plt.plot(df_grand.Year, df_grand.Date, c ='r', label = 'Grand')
other, = plt.plot(df_other.Year, df_other.Date, c = 'g', label = 'Other')
plt.legend(handles =[petty, grand, other], loc = "upper left")

plt.show()
plt.savefig("Fig2_theftbytype.png")
```
